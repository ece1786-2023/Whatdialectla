```python
#import os
from google.colab import userdata
import openai
import pandas as pd
import re
import time
import numpy as np

client = openai.OpenAI(api_key = userdata.get('OPENAI_API_KEY'))
```

```python
dialectDict = {
  "1": "Brt",
  "2": "Sgp",
  "3": "Ind",
  "4": "Aus",
  "5": "Saf",
  "6": "Irs",
  "0": "For"
}
reversedDict = {}
for key, value in dialectDict.items():
    reversedDict[value] = key

print(reversedDict)
```

```python
def classify(statement):
    response = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {
          "role": "system",
          "content": 'You are a dialectologist specializing in English. Your job is to classify the dialect. The classification labels are:\
          1 = British dialect\
          2 = Singapore dialect\
          3 = Indian dialect\
          4 = Australian dialect\
          5 = South African dialect\
          6 = Irish dialect\
          0 = No dialect\
          Just output the labels'
        },
        {
          "role": "user",
          "content": statement
        }
      ],
      temperature=0,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    time.sleep(3)
    label = re.sub("[^0-9]", "", response.choices[0].message.content)
    return label
```

```python
def classify2(statement):
    response = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {
          "role": "system",
          "content": 'You are a dialectologist specializing in English. Your job is to classify the dialect. The classification labels are:\
          1 = British dialect\
          2 = Singapore dialect\
          3 = Indian dialect\
          4 = Australian dialect\
          5 = South African dialect\
          6 = Irish dialect\
          0 = No dialect\
          Here are the sentences classified by English dialects:\
          "Mum said it made her depressed and bitter, and her husband left her because of it." - British English (Classified as 1).\
          "Eh, you very kiasu leh, always must be first in everything." - Singaporean English (Classified as 2).\
          "Side, please. I\'m in a hurry." - Indian English (Classified as 3).\
          "Tim is a bludger; he always tries to sneak his way out of difficult situations." - Australian English (Classified as 4).\
          "I\'ll see you just now; I\'m quickly going to the shop." - South African English (Classified as 5).\
          "We\'re just having a bit of craic at the party." - Irish English (Classified as 6).\
          For general english statements, (Classified as 0).\
          Just output the labels'
        },
        {
          "role": "user",
          "content": statement
        }
      ],
      temperature=0,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    time.sleep(3)
    #print(response.choices[0].message.content)
    label = re.sub("[^0-9]", "", response.choices[0].message.content)
    return label
```

```python
def classify3(statement):
    response = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {
          "role": "system",
          "content": 'You are a professional at classifying English as different dialects. Your job is to classify the dialect. The classification labels are:\
          1 = British dialect \
          2 = Singapore dialect\
          3 = Indian dialect\
          4 = Australian dialect\
          5 = South African dialect\
          6 = Irish dialect\
          0 = No dialectHere\
          A few examples of sentences classified by English dialects:\
          "Mum said it made her depressed and bitter, and her husband left her because of it." - British English (Classified as 1) because Mum is commonly used by a person from britain to address their mother, similarly Bonkers is used to assert than a person is mad or carzy.\
          "He drives so fast, quite scary sometimes." - Singaporean English (Classified as 2) because "quite" coming immediately after a comma is similar to how "sangat" is used in malay and follows the sentence structure of the other languages in the region\
          "Side, please. I\'m in a hurry." - Indian English (Classified as 3) because "Side, please" is very common phrase used in Indian to ask the general public to move out the way in a polite manner. Indian english also end the sentences abruptly such with term such as that\'s all, ok ya, this only no and so on.\
          "Tim is a bludger; he always tries to sneak his way out of difficult situations." - Australian English (Classified as 4) because the word bludger is used to refer to a lazy person, its commonly used by australian\'s.\
          "It\'s a bit chilling today, better than being hot though, isn\'t it?" - South African English (Classified as 5) because South Africans tend to use more double negatives to express themselves, most other dialects would have just use "right?" to end the sentence instead\
          "We\'re just having a bit of craic at the party." - Irish English (Classified as 6) because "criac" is Irish for having a good time and an irish accent tends to be more upbeat.\
          "Our aim is to minimize the effort to train the two-stage pipeline of the conventional teacher-student framework." - General English, (Classified as 0) because there is no specific term or pattern suggesting that it belongs a particular dialect.\
          DO NOT GENERATE ANY NEW CLASSIFICATION BEYOND WHAT\'s DESCRIBED ABOVE'
        },
        {
          "role": "user",
          "content": statement
        }
      ],
      temperature=0,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    time.sleep(3)
    print(response.choices[0].message.content)
    label = re.sub("[^0-9]", "", response.choices[0].message.content)[0]
    return label
```

```python
accuracy = 0
diaAcc = np.zeros((7,7))
```

```python
content = pd.read_csv("datasetWAnno.csv", index_col=None, header=None)

for count, senLabPair in enumerate(content.values):
  label = classify2(senLabPair[0])
  diaAcc[int(reversedDict[senLabPair[1]])][int(label)] += 1
  label = dialectDict[label]
  print(count+1, label, senLabPair[1])
  if label == senLabPair[1]:
    accuracy += 1

print(diaAcc)
accuracy_percentage = accuracy / len(content) * 100
print(accuracy_percentage)
```

```python
accuracy = 0
diaAcc_chain = np.zeros((7,7))
labels=[]
```

```python

content = pd.read_csv("/content/datasetWAnno 1.csv", index_col=None, header=None)
for count, senLabPair in enumerate(content.values):
  if senLabPair[0] not in ['!','?','.']:
    senLabPair[0] += "."
  label = classify3(senLabPair[0])
  label = dialectDict[label]
  labels.append(label)
  print(count+1, label, senLabPair[1])
```

```python
for label, senLabPair in (content.values, labels):
  if label == senLabPair[1]:
      accuracy += 1
      diaAcc_chain[int(reversedDict[senLabPair[1]])][int(label)] += 1

print(diaAcc_chain)
accuracy_percentage = accuracy / len(content) * 100
print(accuracy_percentage)
```
