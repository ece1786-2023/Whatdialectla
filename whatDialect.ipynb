```python
dialectDict = {
  "1": "Brt",
  "2": "Sgp",
  "3": "Ind",
  "4": "Aus",
  "5": "Saf",
  "6": "Irs",
  "0": "For",
  "-1": "Nan",
  1: "Irish",
  2: "South African",
  3: "Singaporean",
  4: "Australian",
  5: "Indian",
  6: "British",
  0: "None",
}
reversedDict = {}
for key, value in dialectDict.items():
    reversedDict[value] = key

print(reversedDict)
```

#ZeroShot

```python
def classifyZS(statement):
    response = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {
          "role": "system",
          "content": 'You are a dialectologist specializing in English. Your job is to classify the dialect. The classification labels are:\
          1 = British dialect\
          2 = Singapore dialect\
          3 = Indian dialect\
          4 = Australian dialect\
          5 = South African dialect\
          6 = Irish dialect\
          0 = No dialect\
          Just output the labels'
        },
        {
          "role": "user",
          "content": statement
        }
      ],
      temperature=0,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    time.sleep(1)
    label = re.sub("[^0-9]", "", response.choices[0].message.content)
    return label
```

```python
accuracy = 0
diaAcc = np.zeros((7,7))
```

```python
content = pd.read_csv("datasetWAnno.csv", index_col=None, header=None)

for count, senLabPair in enumerate(content.values):
  label = classifyZS(senLabPair[0])
  #diaAcc[int(reversedDict[senLabPair[1]])][int(label)] += 1
  label = dialectDict[label]
  print(count+1, label, senLabPair[1])
  if label == senLabPair[1] or label == senLabPair[2]
    accuracy += 1

#print(diaAcc)
accuracy_percentage = accuracy / len(content) * 100
print(accuracy_percentage)
```

#ChainOfThought

```python
def classify3(statement):
    response = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {
          "role": "system",
          "content": 'You are a professional at classifying English as different dialects. Your job is to classify the dialect. The classification labels are:\
          1 = British dialect\
          2 = Singapore dialect\
          3 = Indian dialect\
          4 = Australian dialect\
          5 = South African dialect\
          6 = Irish dialect\
          0 = No dialect\
          A few examples of sentences classified by English dialects:\
          "Mum said it made her depressed and bitter, and her husband left her because of it." - British English (Classified as 1) because Mum is commonly used by a person from britain to address their mother, similarly Bonkers is used to assert than a person is mad or crazy.\
          "He drives so fast, quite scary sometimes." - Singaporean English (Classified as 2) because "quite" coming immediately after a comma is similar to how "sangat" is used in malay and follows the sentence structure of the other languages in the region\
          "Side, please. I\'m in a hurry." - Indian English (Classified as 3) because "Side, please" is very common phrase used in Indian to ask the general public to move out the way in a polite manner. Indian english also end the sentences abruptly such with term such as that\'s all, ok ya, this only no and so on.\
          "Tim is a bludger; he always tries to sneak his way out of difficult situations." - Australian English (Classified as 4) because the word bludger is used to refer to a lazy person, its commonly used by australian\'s.\
          "It\'s a bit chilling today, better than being hot though, isn\'t it?" - South African English (Classified as 5) because South Africans tend to use more double negatives to express themselves, most other dialects would have just use "right?" to end the sentence instead\
          "We\'re just having a bit of craic at the party." - Irish English (Classified as 6) because "criac" is Irish for having a good time and an irish accent tends to be more upbeat.\
          "Our aim is to minimize the effort to train the two-stage pipeline of the conventional teacher-student framework." - General English, (Classified as 0) because there is no specific term or pattern suggesting that it belongs a particular dialect.\
          DO NOT GENERATE ANY NEW CLASSIFICATION BEYOND WHAT\'s DESCRIBED ABOVE'
        },
        {
          "role": "user",
          "content": statement
        }
      ],
      temperature=0,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    time.sleep(1)
    print(response.choices[0].message.content)
    label = re.sub("[^0-9]", "", response.choices[0].message.content) #returns two labels if there are 2, returns 1 if there is only 1
    if len(label) < 2:
      return label
    else:
      return label[0:2]
```

#Zero shot generator

```python
def regenerate(statement, targetDialect):
    response = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {
          "role": "system",
          "content": f"You are a dialectologist specializing in English. Your job is to translate the prompt sentence into {targetDialect} English.",
        },
        {
          "role": "user",
          "content": statement
        }
      ],
      temperature=0,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )

    #print(statement)
    #print(targetDialect+":",response.choices[0].message.content)
    return response.choices[0].message.content
```

```python
csvfile = "genSentences"
content = pd.read_csv(f"/content/{csvfile}.csv", index_col=None, header=None)
responses = []
for count, senLabPair in enumerate(content.values):
    response = regenerate(senLabPair[0], dialectDict[int(map[count])])
    responses.append(response)
    print(count+1, senLabPair[0])
    print(dialectDict[map[count]]+":",response)
```

```python
data = pd.DataFrame(content.values,columns=['Input','Class_I'])
data['Output'] = responses
data['Class_O'] = tag
data.to_csv("Outputfile.csv")
```

#Few Shot Generator

```python
exampleSentences = ["","","'I tried my level best to not be late da, but the traffic here is really bad'",
                    "'Bro, I tried to be on time, but the jam on the way here really cannot.'",
                    "'I usually make an effort to be punctual, but the traffic was absolutely bollocks pal.'",
                    "'Mate, I tried to be here on time, but the traffic was shocking today.'"]
```

```python
def regenerateFS(statement, targetDialect, exampleSentence):
    response = client.chat.completions.create(
      model="gpt-4",
      messages=[
        {
          "role": "system",
          "content": f"You are a dialectologist specializing in English. Your job is to translate the prompt sentence into {targetDialect} English.\
          Example, translating this sentences: 'Dude, I tried to be here on time, but the traffic is horrible.' to {exampleSentence}",
        },
        {
          "role": "user",
          "content": statement
        }
      ],
      temperature=1,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )

    #print(statement)
    #print(targetDialect+":",response.choices[0].message.content)
    return response.choices[0].message.content
```

```python
map = [3, 3, 4, 4, 5, 5, 6, 6,
       3, 3, 4, 4, 5, 5, 6, 6,
       4, 4, 4, 5, 5, 5, 6, 6, 6,
       3, 3, 3, 5, 5, 5, 6, 6, 6,
       3, 3, 3, 4, 4, 4, 6, 6, 6,
       3, 3, 3, 4, 4, 4, 5, 5, 5,]
```

```python
csvfile = "genSentencesAlt"
content = pd.read_csv(f"/content/{csvfile}.csv", index_col=None, header=None)
responses,tag = [],[]
for count, senLabPair in enumerate(content.values):
  targetDialect = dialectDict[int(map[count])]
  exampleSentence = exampleSentences[int(map[count])-1]
  response = regenerateFS(senLabPair[0], targetDialect,exampleSentence)
  responses.append(response)
  tag.append(dialectDict[int(map[count])])
  print(count+1, senLabPair[0])
  print(dialectDict[map[count]]+":",response)
```

```python
data = pd.DataFrame(content.values,columns=['Input','Class_I'])
data['Output'] = responses
data['Class_O'] = tag
data.to_csv("Outputfile.csv")
```
